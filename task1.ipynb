```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
# drop id from dataset
columns_to_drop = ['id', 'Unnamed: 32']
drop_id = data.drop(columns=columns_to_drop, axis=1) # drop id column from dataset 
drop_id.info()
```

```python
missing = data.isnull().sum().sort_values(ascending= False)
missing_per = (data.isnull().sum() / data.isnull().count()).sort_values(ascending= False)
total = pd.concat([missing, missing_per], axis=1, keys=['Total', \
                                                        'Percent'])
total
```

```python
# Check whether the dataset is an imbalanced dataset

y = drop_id['diagnosis']
from collections import Counter
scope_diagnosis = Counter(y)
print(scope_diagnosis)
```

```python
# Separate data based on the class
data_0 = drop_id[drop_id['diagnosis'] == 'M']
data_1 = drop_id[drop_id['diagnosis'] == 'B']
```

```python
# Randomly undersample the majority class

arr = data_1.index 
rand_num = np.random.permutation(arr).reshape((len(arr), 1))
data_1.drop(index=rand_num[0:145,0], inplace=True)
```

```python
# Combine data_0 and data_1

balanced_data = pd.concat([data_0, data_1], ignore_index=True)
```

```python
# Separate the balanced data into features and target

x = balanced_data.drop('diagnosis', axis=1)
y = balanced_data['diagnosis']

balanced_data.head()
scope_diagnosis = Counter(y)
print(scope_diagnosis)
```

```python
# Input normalization

from sklearn.preprocessing import StandardScaler

#Normalize the feature sets

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
len(x_test_scaled)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier

models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "AdaBoost": AdaBoostClassifier(),
    "SVM": SVC(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB(),
    "Neural Network": MLPClassifier()
}

# Train network
'''
    เป็นการทดสอบใช้ในทุก Model ที่กล่าวไปใน models ลองทำทุกรูปแบบที่เป็นไปได้

'''
for name, model in models.items():
    model.fit(x_train_scaled, y_train)


c_chart_mat = []
class_names = list(data['diagnosis'].unique())

from sklearn.metrics import confusion_matrix
for name, model in models.items():
    y_pred = model.predict(x_train_scaled)
    # Calculate the confusion matrix 
    c_chart = confusion_matrix(y_train, y_pred)
    c_chart_mat.append(c_chart)

    print("$",name)
    print("c_chart=", c_chart)
    # Visualize the confusion matrix
    import seaborn as sns

    sns.heatmap(c_chart, annot=True, cmap='Blues', fmt='d',xticklabels=class_names)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```

```python
#compute performance metrices from the confusion matrices
# Accuracy , Precison, Recall, F1-score, specificity, sensitivity

# performance parameters of the training dataset
#P1 : Accuracy = (TP+TN)/(TP+FN+TN+FP)
#P2 : Specificity = TN / (TN+FP)
#P3 : Precision = TP / (TP + FP)
#P4 : Recall = Sensitivity = TP / (TP + FN)
#P5 : F1_score = 2 * Precision * Recall / (Precision + Recall)

from sklearn.metrics import confusion_matrix, precision_recall_fscore_support 
for name, model in models.items():
    y_pred = model.predict(x_train_scaled)
    c_chart = confusion_matrix(y_train, y_pred)

    # Extract

    TN, FP, FN, TP = c_chart.ravel()

    # Total actual positives and negatives

    total_positives = TP + FN
    total_negatives = TN + FP 

    # Normalizing the confusion matrix components

    TP = TP / total_positives
    FN = FN / total_positives
    TN = TN / total_negatives
    FP = FP / total_negatives

    # Calculating Accuracy , Precision , Recall , F1-Score

    accuracy = ( TP + TN) / (TP + FP + FN + TN) # as a parameter 1
    specificity = TN / (TN + FP) # as a parameter 2
    precision = TP / (TP + FP) # as a parameter 3
    recall = TP / (TP + FN) # same as sensitivity
    f1_score =  2 * (precision * recall) /  (precision + recall)

    # presentation information

    print(f' ----------------  {name} : Traning dataset  ----------------')
    print(f"confusion_matrix: {c_chart}")
    print(f'Accuracy: {accuracy:.5f}')
    print(f'Precision: {precision:.5f}')
    print(f'Recall: {recall:.5f}')
    print(f'F1-Score: {f1_score:.5f}')
    print(f'Specificity: {specificity:.5f}\n')
```

```python
# Receiver Operating Characteristic (ROC) 
'''
1: Prefect classifier
0.5: No discriminative power
Below 0.5: Worse than random

'''
from sklearn.metrics import roc_curve, auc 

for name, model in models.items():
    plt.figure()

    if hasattr(model, "predict_proba"):
        probas = model.predict_proba(x_train_scaled)
        y_scores = probas[:, 1]

    else:
        y_scores = model.decision_function(x_train_scaled)
    # Compute ROC curve and AUC
    y_train.dropna()
    #y_train= y_train.map({'B': 0, 'M': 1}).astype(int)
    fpr, tpr, _ = roc_curve(y_train, y_scores)
    roc_auc = auc(fpr, tpr)
        
    plt.plot(fpr, tpr, lw=.8, label=f'ROC curve (area = {roc_auc:.5f})')
    plt.plot([0, 1], [0, 1], 'k--', lw=.8)
    plt.xlim([-0.05, 1.05])
    plt.ylim([-0.05, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel(f'Receiver Operating-Characteristic of {name}')
    plt.legend(loc="lower right")
    plt.show()
```
